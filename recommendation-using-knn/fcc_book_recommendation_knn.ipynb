```python
book_count = df_ratings['isbn'].value_counts().reset_index()
user_count = df_ratings['user'].value_counts().reset_index()

df_ratings_cleaned = df_ratings[df_ratings['user'].isin(user_count[user_count['count'] >= 200]['user'])]
df_ratings_cleaned = df_ratings_cleaned[df_ratings_cleaned['isbn'].isin(book_count[book_count['count'] >= 100]['isbn'])]

df_merge = pd.merge(df_books, df_ratings_cleaned, on='isbn', how='inner')
```

```python
df_merge = df_merge.drop_duplicates(subset=['title', 'user'], keep='first')
df_merge_pivot = df_merge.pivot(index='title', columns='user', values='rating').fillna(0)
df_merge_sp = csr_matrix(df_merge_pivot.values)
```

```python
df_merge_pivot
```

```python
model = NearestNeighbors(metric= "cosine").fit(df_merge_sp)
```

```python
# function to return recommended books - this will be tested
def get_recommends(book = ""):
  recommended_books = coordinate = []
  distance, indexes = model.kneighbors(df_merge_pivot.iloc[df_merge_pivot.index== book].values.reshape(1,-1), n_neighbors=6)
  #print(df_merge_pivot.iloc[indexes[0][5]].name)
  #print(len(distance[0]), len(indexes[0]))
  for i in range(len(distance[0])):
    if i != 0:
      coordinate.append([df_merge_pivot.iloc[indexes[0][i]].name, distance[0][i]])
  coordinate.reverse()
  recommended_books = [book, coordinate]
  return recommended_books

get_recommends(book="The Queen of the Damned (Vampire Chronicles (Paperback))")
```

```python
# --- Provided by freeCodeCamp ---
books = get_recommends("Where the Heart Is (Oprah's Book Club (Paperback))")
print(books)

def test_book_recommendation():
  test_pass = True
  recommends = get_recommends("Where the Heart Is (Oprah's Book Club (Paperback))")
  if recommends[0] != "Where the Heart Is (Oprah's Book Club (Paperback))":
    test_pass = False
  recommended_books = ["I'll Be Seeing You", 'The Weight of Water', 'The Surgeon', 'I Know This Much Is True']
  recommended_books_dist = [0.8, 0.77, 0.77, 0.77]
  for i in range(2):
    if recommends[1][i][0] not in recommended_books:
      test_pass = False
    if abs(recommends[1][i][1] - recommended_books_dist[i]) >= 0.05:
      test_pass = False
  if test_pass:
    print("You passed the challenge! 🎉🎉🎉🎉🎉")
  else:
    print("You haven't passed yet. Keep trying!")

test_book_recommendation()
```
